```python
import sys
import os
sys.path.append("../")
from datapath import base_path, data_path
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
weather = pd.read_csv(os.path.join(data_path, "weather.csv"))
```

```python
weather.head()
```

```python
print(weather.iloc[:5, :])
```

```python
weather_melt = pd.melt(weather, id_vars=["id", "year", "month", "element"], var_name="day", value_name="temp")
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(
    index = ["id", "year", "month", "day"],
    columns="element",
    values="temp",
    dropna=False
)
print(weather_tidy)
```

```python
weather_tidy_flat = weather_tidy.reset_index()
print(weather_tidy_flat.head())
```

```python
import numpy as np
np.count_nonzero(weather_tidy_flat.isnull())
```

```python
weather_tidy_flat.month.value_counts(dropna=False)
```

```python
weather_tidy_flat["datetime"] = weather_tidy_flat["year"].astype(str) + "-" + weather_tidy_flat["month"].astype(str) + "-" + weather_tidy_flat["day"].str[1:]
```
